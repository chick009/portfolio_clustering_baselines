```python
data = pd.read_csv('result_df_SP500_v2.csv', index_col = [0, 1, 2, 3, 4, 5])
data = data.dropna(how = 'all')
```

```python
data["Sharpe Ratio"] = data["Sharpe Ratio"].clip(lower=-3, upper=3)
data["Current Sharpe Ratio"] = data["Current Sharpe Ratio"].clip(lower=-3, upper=3)
data["consistency_ratio"] =  data['Sharpe Ratio'] / data['Current Sharpe Ratio']

for idx, row in data.iterrows():
    data.loc[idx, 'year'] = idx[1][-4:]
```

```python
import numpy as np

result_df_1 = pd.DataFrame(index = data['year'].unique(), columns=['1M', '3M', '6M', '1Y'])
result_df_2 = pd.DataFrame(index = data['year'].unique(), columns=['1M', '3M', '6M', '1Y'])

for interval in ['1M', '3M', '6M', '1Y']:
    # Get the equally rated dataframe
    ae_arr = data.loc[(interval, slice(None), 'yes', slice(None), slice(None), slice(None))]['Sharpe Ratio'].values
    nae_arr = data.loc[(interval, slice(None), 'no', slice(None), slice(None), slice(None))]['Sharpe Ratio'].values
    
    # Calculate the absolute error between sharpe ratio of EW and MVO portfolio
    abs_err = sum(ae_arr - nae_arr) / len(nae_arr)
    
    # Percentage of EW has greater sharpe ratio 
    comparison = ae_arr >= nae_arr
    count_greater = np.count_nonzero(comparison)
    pct_ew_over = (count_greater / len(ae_arr)) * 100

    # Prepare a result dataframe with 
    print(interval, abs_err, pct_ew_over)

    for year in data['year'].unique():
        ae_df_year = data.loc[(interval, slice(None), 'yes', slice(None), slice(None), slice(None))]
        nae_df_year = data.loc[(interval, slice(None), 'no', slice(None), slice(None), slice(None))]
        
        ae_arr_year = ae_df_year[ae_df_year["year"] == year]["Sharpe Ratio"].values
        nae_arr_year = nae_df_year[nae_df_year["year"] == year]["Sharpe Ratio"].values
        
        if len(ae_arr_year) == 0:
            continue

        abs_err = sum(ae_arr_year - nae_arr_year) / len(ae_arr_year)
    
        result_df_1.loc[year, interval] = abs_err

        comparison = ae_arr_year > nae_arr_year
        count_greater = np.count_nonzero(comparison)
        
        pct_ae_over = (count_greater / len(ae_arr_year)) * 100
        result_df_2.loc[year, interval] = pct_ae_over

result_df_1.to_csv('ae_or_not_compar_1.csv')
result_df_2.to_csv('ae_or_not_compar_2.csv')
print("\n", result_df_1, "\n")
print(result_df_2) 
        

```
